### **5. Generar los numeros aleatorios**

```python
import numpy as np #llamada de librerias que se utilizaran para resolver el problema
import random 
import matplotlib.pyplot as plt 
```

```python
def hrs_averias(): 
    hrs=0
    v = random.random() #Se generan números aleatorios
    if 0.0 <= v < 0.0499:
        v = 20
    elif 0.05 <= v < 0.199: #Dependiendo del número, va a obtener un valor
        v = 30
    elif 0.20 <= v < 0.3499:
        v = 40
    elif 0.35 <= v < 0.5499:
        v = 50
    elif 0.55 <= v <= 0.7499:
        v = 60
    elif 0.75 <= v < 0.8999:
        v = 70
    elif 0.90 <= v <= 1.00:
        v = 80
    return v #Devolvemos el valor que el número aleatorio obtuvo
```

###**6. Gráficamos**

```python
def con1(n): #politica 1
  reemplazo=0 #Variable para la suma de las horas entre averias
  costo=[] #Variable para guardar el coste de funcionamiento
  suma_coste=0  
  for i in range(n):
    suma_coste=suma_coste+hrs_averias() 
    reemplazo=reemplazo+110
    costo.append(reemplazo/suma_coste) 

  return costo #Devolvemos el coste de funcionamiento
```

```python
X=con1(50)
```

```python
plt.plot(X, color="green") #Graficamos
```

```python
np.mean(X) 
```

###**7. Replicar el modelo $n$ veces**

```python
def modelo(n): 
    lista=[]
    for i in range(n):
        s = con1(500)
        lista.append(np.mean(s)) 
    return lista
```

```python
mod=modelo(50) #Graficamos la repeticion del modelo
```

```python
print(np.mean(mod), np.std(mod))
```

```python
plt.hist(mod, color="pink") 
plt.show()
```

###**7. Encontrar el intervalo de confianza**

```python

int1= np.mean(mod)-(1.96)*(np.var(mod))/np.sqrt(50)
int2= np.mean(mod)+(1.96)*(np.var(mod))/np.sqrt(50) 

print("El intervalo de confianza es: ",int1, "<",np.mean(mod),"<",int2)
print("es decir, ")
print("El intervalo de confianza es: ",
np.mean(mod)-(1.96)*((np.var(mod))/np.sqrt(50)), "≤ X̅ ≤",
np.mean(mod)+(1.96)*((np.var(mod))/np.sqrt(50)))
```
